```python
# *****************************************************************************************
# relation between diabetic person turning for appointment
# *****************************************************************************************
size = med_app_data.groupby(['Diabetes','No-show']).size()
print ("size : \n",size )
stack_size = med_app_data.groupby(['Diabetes','No-show']).size().unstack()
print ("\n\n size in stack format : \n",stack_size )
med_app_data.groupby("No-show").hist(column="Diabetes")

group_carriers = med_app_data.groupby('No-show-bool')['Diabetes'] # group values by carrier, select minutes delayed
print("group_carriers =",group_carriers)
group_carriers.plot(kind='hist', figsize=[12,6], alpha=.4, legend=True) # alpha for transparency
```

```python
# *****************************************************************************************
# relation between Gender & attending appointment
# *****************************************************************************************
size = med_app_data.groupby(['Gender','No-show']).size()
print ("size : \n",size )
stack_size = med_app_data.groupby(['Gender','No-show']).size().unstack()
print ("\n\n size in stack format : \n",stack_size )
```
