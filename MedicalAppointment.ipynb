```python
med_app_data['Scheduled_hour'] =  (med_app_data.ScheduledDay.apply(lambda x:x.hour))
```

```python
def get_day(datetime_obj):
    return datetime_obj.date()
```

```python
med_app_data['lead_days'] = (med_app_data.AppointmentDay.apply(get_day) - 
                             med_app_data.ScheduledDay.apply(get_day))
```

```python
med_app_data.iloc[0]
```

```python
med_app_data.lead_days = (med_app_data.lead_days.astype("timedelta64[D]")).astype(int)
```

```python
med_app_data.iloc[0]
```

```python
med_app_data.dtypes
```

```python
med_app_data['appointment_DOW'] = (med_app_data.AppointmentDay.dt.dayofweek)
```

```python
med_app_data.iloc[0]
```

```python
med_app_data['lead_days'].describe()
```

```python
# create lead_days_category column
lead_days_labels = pd.Series(['A:Same Day','B:1-2 Days','C:3-7 Days','D:8-31 Days','E: 32+Days'])
```

```python
med_app_data['lead_days_category'] = pd.cut(med_app_data.lead_days, bins = [-1,0,2,7,31,999], labels= lead_days_labels)
```
